```python
#bag of features
#length
#
#
#

import string
characters = string.printable
MAX_LEN = 10


def flat(a):
    return np.array(a).reshape(-1)

def mmm(seq): # where mmm stands for min, mean, 
    return np.array([seq.min(), seq.mean(), seq.max()])

def to_str(seq):
    return seq.map(lambda x: str(x))



def length(seq):
    lengths = seq.map(lambda s: len(s))
    return mmm(lengths)

def character_appears(seq):
    features = []
    for character in characters:
        occurencies = seq.map(lambda s: s.count(character))
        features.append(mmm(occurencies))
    return flat(features)
        
    
def fixed_position_character_appears(seq):
    occurencies = np.zeros((len(characters)*2, MAX_LEN))
    for character in characters:
        for s in seq:
            for i, c in enumerate(s):
                if (i == MAX_LEN):
                    break
                occurencies[ord(character)][i] += (c == character)
    return flat(occurencies)

def
    
def bag_of_features(seq):
    detectors = [
        length,
        character_appears,
        fixed_position_character_appears,
    ]
    
    features = np.array([])
    for detector in detectors:
        features = np.append(detector(to_str(seq)), features)
    return features
    
```
